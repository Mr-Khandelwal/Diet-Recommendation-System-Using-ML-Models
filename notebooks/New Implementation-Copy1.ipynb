## Removing the outliers


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Plot boxplots for numeric columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.show()
```

- The central box represents the interquartile range (IQR), which contains the middle 50% of the data.
- The bottom of the box represents the first quartile (Q1), also known as the 25th percentile.
- The top of the box represents the third quartile (Q3), also known as the 75th percentile.
- The line inside the box represents the median (Q2 or the 50th percentile) of the data.

- The "whiskers" extend from the box to the smallest and largest values within a certain range. Typically, the whiskers extend to 1.5 times the IQR from the Q1 and Q3.
- The lower whisker extends to the smallest data point within 1.5 times the interquartile range (IQR) below the first quartile (Q1).
- The upper whisker extends to the largest data point within 1.5 times the IQR above the third quartile (Q3).

```python
numerical_columns = df.select_dtypes(include=(np.number)).columns

def remove_outliers(df, column):
    lower = df[column].quantile(0.25)
    upper = df[column].quantile(0.75)
    IQR = upper - lower
    lower_whisker = lower - IQR * 1.45
    upper_whisker = upper + IQR * 1.45
    df = df[(df[column] > lower_whisker) & (df[column] < upper_whisker)]
    return df

# Apply the remove_outliers function only to numerical columns
for column in numerical_columns:
    df = remove_outliers(df, column)
```

```python
6# Function to calculate BMR
def calculate_bmr(weight, height, age, gender):
    if gender == 1:  # 0 represents Male
        bmr = 447.593 + (9.247 * weight) + (3.098 * height) - (4.330 * age)
    elif gender == 0:  # 1 represents Female
        bmr = 88.362 + (13.397 * weight) + (4.799 * height) - (5.677 * age)
    else:
        raise ValueError("Invalid gender. Use '0' for Male or '1' for Female.")
    return bmr

# Function to calculate Body Fat Percentage
def calculate_body_fat_percentage(bmi, age, gender):
    if age >= 18 and age <= 21:
        return (1.51 * bmi) - (0.70 * age) - (3.6 * gender) + 14
    elif age >= 22:
        return (1.391 * bmi) + (0.16 * age) - (10.34 * gender) - 9
    else:
        raise ValueError("Invalid age range")

# Convert relevant inputs to appropriate types
gender = int(inputs[0])
age = int(inputs[1])
height = float(inputs[2])
weight = float(inputs[3])

# Calculate BMI
bmi = weight / ((height / 100) ** 2)
inputs.append(bmi)
print("BMI: ", bmi)

# Calculate BMR
bmr = calculate_bmr(weight, height, age, gender)
inputs.append(bmr)
print("BMR: ", bmr)

# Calculate Body Fat Percentage
body_fat_percentage = calculate_body_fat_percentage(bmi, age, gender)
inputs.append(body_fat_percentage)
print("BFP: ", body_fat_percentage)
```
